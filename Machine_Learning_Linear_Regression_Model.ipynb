# Machine Learning Example
## Linear Regression with Python

***

This is going to use a small data set for simplicity

```python
# Import all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Create data frame
df = pd.DataFrame({
    'X': [1,2,3,4,5],
    'Y': [1.2, 2.3, 2.8, 3.1, 3.9]
})

# View data frame
df.head()
```

## Checking to see if linear model is appropriate

It is important to plot a regplot and residplot to check if the a linear regression model is the correct choice.

```python
# Plotting regplot
sns.regplot(x='X', y='Y', data=df)
plt.title('Linear Regression Plot')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
sns.residplot(df['Y'], x = df['X'], y = df['Y'])
plt.title('Residual Plot')
```

The regplot seems to match up well, and the residplot shows randomly spread out data points. This is what we want.
***
Now, we build and train the linear regression model.

```python
# Build the model
model = LinearRegression()
```

```python
# Train the model
model.fit(df[['X']], df['Y'])
```

```python
# Make predictions
y_pred = model.predict(df[['X']])
```

```python
# Validate the model
mse = mean_squared_error(df['Y'], y_pred)
r2 = r2_score(df['Y'], y_pred)

# Print evaluation metrics
print(f"Mean Squared Error: {mse}")
print(f"R-squared Error: {r2}")
```

***
Mean Squared Error is low and R-squared is close to 1. It seems we have a good-fitting model.
***

```python
# Plot the a scatter plot of the original x and y data
plt.figure(figsize=(8, 6))
plt.scatter(x = df['X'], y = df['Y'], color = 'black', label = 'Data')

# Plot your new regression line with your predictions
plt.plot(df['X'], y_pred, color = 'red', label = 'Predictions')

# Customize however you would like
plt.title("Simple Linear Regression Model Example")
plt.xlabel('The X Data')
plt.ylabel('The Y Predictions')
plt.xticks(df['X'])
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.text(0, -1,
         s = f"Mean Squared Error: {mse}\nR-squared: {r2}",
         fontsize = 8,
         bbox = dict(facecolor = 'white', edgecolor = 'black'))
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```
